```python
#Dependencies 
import time
from splinter import Browser
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
from selenium import webdriver
```

```python
#URL's of pages to be scraped
nasa_url = 'https://mars.nasa.gov/news/'
jpl_url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
```

```python
# Retrieve NASA page with the requests module
response = requests.get(nasa_url)

#Display information from page being scraped
print(response.text)
```

```python
#Create BeautifulSoup object; parse with 'html' or 'lxml'
soup = bs(response.text, 'html.parser')
```

```python
#Retrieve the parent divs for all NASA articles
results = soup.find_all('div', class_='slide')

#Print results
print(results)
```

```python
# Loop through results to retrieve all newstitle and all news paragraph to store in MongoDB for future use
for result in results:
    newstitle = result.find('div', class_='content_title').text

    newsp = result.find('div', class_='rollover_description_inner').text
    print('-----------------')
    print(newstitle)
    print(newsp)

    # Dictionary to be inserted into MongoDB
    post = {
        'News_Title': newstitle,
        'News_Paragraph': newsp,
    }
```

```python
#Retrieve the title for all NASA articles
titles = soup.find_all('div', class_='content_title')   
print(titles)
```

```python
#Print the most recent article title
newstitle = []
for title in titles:
    newstitle.append(title)

for x in range(1):
    news_t = (newstitle[x].text)
    print(news_t)
```

```python
#Retrieve the news paragraph for all NASA articles
paragraphs = soup.find_all('div', class_='rollover_description_inner')   
print(paragraphs)
```

```python
#Print the most recent article paragraph
news_paragraph = []
for news in paragraphs:
    news_paragraph.append(news)

for x in range(1):
    news_p = (news_paragraph[x].text)
    print(news_p)
```
