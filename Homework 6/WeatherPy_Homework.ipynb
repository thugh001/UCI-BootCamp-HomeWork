```python
# Dependencies
import requests
import json
import numpy as np
import pandas as pd
from citipy import citipy
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate random list of 500+ Cities 
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

print(lat_lngs)
```

```python
# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
   # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the number of cities that generated in random data set
len(cities)
```

```python
# Create data frame that will hold random city names and corresponding data
weather_df = pd.DataFrame({'City': cities})
weather_df['Latitude'] = ""
weather_df['Temperature'] = ""
weather_df['Humidity'] = ""
weather_df['Cloudiness'] = ""
weather_df['Wind Speed'] = ""
weather_df.head()
```

```python
# Save config information.
api_key = "632236eb6405d98f1b3fd97eefaaa800"
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
```

```python
# Create empty list to store data
city_list = []

# Loop through cities in weather dataframe and perform api calls to gather data
for index, row in weather_df.iterrows():
    city = row["City"]
    target_url = "http://api.openweathermap.org/data/2.5/weather?units=imperial&q=%s&appid=%s" % (city.replace(" ", "+"), api_key)
    weather = requests.get(target_url).json()
    print(city + "\n" + target_url)
    
    # retrieving all the fields required in assignment
    try:
        weather_df.set_value(index, "Latitude", weather['coord']['lat'])
        weather_df.set_value(index, "Temperature", weather['main']['temp_max'])
        weather_df.set_value(index, "Humidity", weather['main']['humidity'])
        weather_df.set_value(index, "Cloudiness", weather['clouds']['all'])
        weather_df.set_value(index, "Wind Speed", weather['wind']['speed'])
        city_list.append(row['City'])
    except:                                                           
        print(" Missing field... skipping.")
                
        
# Remove rows from dataframe that have missing data 
weather_df = weather_df[weather_df['City'].isin(city_list)] 

weather_df.head()                                                                                                                              
```
